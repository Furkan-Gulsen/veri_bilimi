```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
tips.head(10)
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x="day",y="total_bill",data=tips)
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x="day",y="total_bill",hue="sex",data=tips)
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x="tip",y="day",data=tips)
```

```python
# order: sıralama
plt.figure(figsize=(15,6))
ax = sns.barplot(x="time",y="tip",data=tips,order=["Dinner","Lunch"])
```

```python
# mediana göre grafikleme
from numpy import median
plt.figure(figsize=(15,6))
ax = sns.barplot(x="day",y="tip",data=tips,estimator=median)
```

```python
# hata çubuğu
plt.figure(figsize=(15,6))
ax = sns.barplot(x="day", y="tip", data=tips, ci=70)
```

```python
# güven aralığı
plt.figure(figsize=(15,6))
ax = sns.barplot(x="day", y="tip", data=tips, ci="sd")
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x="day", y="tip", data=tips, capsize=.25)
```

```python
# renk geçişleri
plt.figure(figsize=(15,6))
ax = sns.barplot(x="size",y="total_bill",data=tips,palette="Blues")
```

```python
# Çubuk konumunu veya genişliğini değiştirmeden ton kullanma
plt.figure(figsize=(15,6))
tips["weekend"] = tips["day"].isin(["Sat","Sun"])
ax = sns.barplot(x="day",y="total_bill",hue="weekend",data=tips,dodge=False)
```

```python
# tek renk verme
plt.figure(figsize=(15,6))
ax = sns.barplot(x="size",y="total_bill",data=tips,color="salmon",saturation=.5)
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot("day", "total_bill", data=tips,
                 linewidth=2.5, facecolor=(1, 1, 1, 0),
                 errcolor=".2", edgecolor=".2")
```
