## K Fold Cross Validation
- Avoid Overfittingi engeller.
- Train olan datayı K kere böl.

```python
from IPython.display import Image
Image(url="./img/model_selectio1.jpg")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
# 4 feature var, 3 class var(target)
iris = load_iris()
data = iris.data
feature_names = iris.feature_names
x = iris.data
y = iris.target

df = pd.DataFrame(data,columns=feature_names)
df
```

```python
# normalization
x = (x-np.min(x))/(np.max(x)-np.min(x))
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3) # key değeri
```

```python
# key fold CV K = 10
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=knn,X=x_train,y=y_train,cv=10) # k kadar bölmek

accuracies
```

```python
print("Acerage Accuracy = ", np.mean(accuracies)*100)
print("Acerage STD = ", np.std(accuracies)*100)
```

```python
knn.fit(x_train,y_train)
print("Test Accuracy = ",knn.score(x_test,y_test)*100)
```

## Grid Search Cross Validation

```python
Image(url="./img/grid_search.jpg")
```

```python
# en uygun n_neighbors değerini bulmamı sağlar
from sklearn.model_selection import GridSearchCV

grid = {"n_neighbors":np.arange(1,50)}
knn = KNeighborsClassifier()

knn_cv = GridSearchCV(knn,grid,cv=10)
knn_cv.fit(x,y)

# print hyperparameter KNN Algoritmasındaki K değeri
print("Tuned Hyperparameter K: ",knn_cv.best_params_)
print("Tuned parametreye göre en iyi accuracy(best score): ", knn_cv.best_score_)
```
