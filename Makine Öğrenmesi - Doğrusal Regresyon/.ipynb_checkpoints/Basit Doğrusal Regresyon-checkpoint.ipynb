Statsmodels İle Modelleme

```python
import statsmodels.api as sm

X = df[["TV"]]
X[0:5]
```

```python
X = sm.add_constant(X)
X[0:5]
```

```python
y = df["sales"]
y[0:5]
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()
model.summary()
```

F-statistic => Modelin anlamlı olup olmadığını verir.
